**Question-2:** Find log-likelihood values for 50, 75, and 80. For

California_housing [â€˜latitudeâ€™] and find the Max likelihood values.

```python
def mle_cal(x,mean,var):
  mle = 1/(math.sqrt(2*math.pi*var)) * math.exp(-(x-mean)*2/(2*var))
  return mle
def lle_cal(x,mean,var):
  lle = -0.5*math.log(2*math.pi*var) - ((x-mean)**2/(2*var))
  return lle
```

```python
values = [50,75,80]
mean,var = mean_var(data['latitude'])
n = len(data['latitude'])
for x in values:
  mle = mle_cal(x,mean,var)
  lle = lle_cal(x,mean,var)
  print(f"MLE of {x}: {mle}")
  print(f"LLE of {x}: {lle}\n")
```
